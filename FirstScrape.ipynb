```python
import nltk
nltk.download('punkt')
```

Sentence Tokenization

```python
from nltk.tokenize import sent_tokenize

sent_tokenized_article = sent_tokenize(article)

print(sent_tokenized_article)
```

Word Tokenization

```python
from nltk.tokenize import word_tokenize

#solving hyphenation issue here
article = article.replace("-", " ")

#solving continuation of paras issue here
article = article.replace(".", " ")

word_tokenized_article = word_tokenize(article)

tokens = word_tokenized_article

print(tokens)
```

To lowercase

```python
new_tokens = []
for token in tokens:
    new_token = token.lower()
    new_tokens.append(new_token)
tokens = new_tokens
print(tokens)

```

Punctuation removal

```python
import re
new_tokens = []
for token in tokens:
    new_token = re.sub(r'[^\w\s]', '', token)
    if new_token != '':
        new_tokens.append(new_token)
tokens = new_tokens
print(tokens)
```

Stopword Removal

```python
nltk.download('stopwords')

from nltk.corpus import stopwords

stopwords.words('english')

new_tokens = []
for token in tokens:
    if token not in stopwords.words('english'):
        new_tokens.append(token)
tokens = new_tokens
print(tokens)
```

Lemmatization

```python
#import nltk
#do download all once
#nltk.download('all')
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
lemmas = []
for token in tokens:
    lemma = lemmatizer.lemmatize(token, pos='v')
    lemmas.append(lemma)
tokens = lemmas
print(tokens)
```

Frequency Distribution

```python
freq = nltk.FreqDist(tokens)

for key,val in freq.items():

    print (str(key) + ':' + str(val))
```
